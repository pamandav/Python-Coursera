```python
#Training and Testing Data
#to split the input into training and tetsing sets we use scikit-learn's train_test_split function
#This function extracts 75% - training data. and 25% - test data
#shuffles the dataset using a pseudorandom number generator
# to make sure that we get the same o/p if we run the same function several times, we provide a fixed seed using random_state
# This makes the o/p deterministic.

from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test = train_test_split(iris_dataset['data'],iris_dataset['target'], random_state =0)
```

```python
#FIRST LOOK AT THE DATA
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import pandas.tools.plotting
```

```python

iris_df = pd.DataFrame(X_train, columns = iris_dataset.feature_names)
pd.tools.plotting.scatter_matrix(iris_df, c= y_train, figsize= (15,15), s = 60, marker = 'o')
```
